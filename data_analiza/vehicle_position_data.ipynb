```python
trip_entities.head(10)
```

entity_id -> Unikalny identyfikator encji w feedzie GTFS-RT (konkretny pojazd lub kurs).

is_deleted -> Informacja, czy encja została oznaczona jako usunięta z feedu (zwykle False).

has_trip_update -> Czy encja zawiera dane aktualizacji rozkładu (ETA, opóźnienia).

has_vehicle_position ->Czy encja zawiera dane o pozycji pojazdu (GPS).

has_alert -> Czy encja zawiera powiązane alerty (np. objazdy, komunikaty).

tu_trip_id -> ID kursu w rozkładzie jazdy (powiązany z trip_id z GTFS static).

tu_route_id -> Numer linii, której dotyczy aktualizacja (np. 15, 54B, Z3).

tu_direction_id -> Kierunek jazdy:
                        0 = pierwszy kierunek,
                        1 = drugi kierunek (GTFS: inbound/outbound).

tu_start_time -> Planowana godzina rozpoczęcia kursu z GTFS static.

tu_vehicle_id -> Identyfikator pojazdu wykonującego kurs (numer taborowy).

tu_stop_updates_count -> Liczba aktualizacji przystanków w tym kursie (ile stopów dostało informację o opóźnieniach).

```python
trip_positions.head(10)
```

entity_id -> ID encji, pozwala połączyć z trip_updates_entities.csv.

trip_id -> ID kursu z GTFS (pozwala powiązać z rozkładem).

route_id -> Numer linii, której dotyczy aktualizacja (np. 15, 55, 80A).

stop_id -> ID przystanku dla którego podano aktualizację.

stop_sequence -> Numer przystanku w kolejności na trasie (1 = pierwszy).

arrival_delay -> Opóźnienie przyjazdu w sekundach (np. 120 = 2 min spóźnienia).

arrival_time -> Przewidywana godzina przyjazdu (epoch timestamp).

departure_delay ->Opóźnienie odjazdu w sekundach.

departure_time -> Przewidywana godzina odjazdu (epoch timestamp).

schedule_relationship - > Relacja z rozkładem.
                            Najczęstsze wartości GTFS:
                            •	SCHEDULED — normalny kurs
                            •	SKIPPED — przystanek pominięty
                            •	NO_DATA — brak danych
                            •	ADDED — dodatkowy kurs
